# Loop para calcular el P value para cada GO term

```python
## Loop para calcular el P value para cada GO term
p_val = []
fold = []
for index, row in estadistica.iterrows():
    b=hypergeom.sf(row.list_count - 1, row.tot_back, row.back_count, row.tot_list, loc=0)
    p_val.append(b)
    fold_enr = (row.list_count / row.tot_list) / (row.back_count / row.tot_back)
    fold.append(fold_enr)
estadistica['P'] = p_val
estadistica['FE'] = fold
```

```python
"""
En el caso de GO terms con un gen no es posible medir la sobrerreprentación de términos
por lo tanto se son removidos
"""
estadistica2 = estadistica[estadistica.back_count > 1].reset_index(drop=True)
```

# Loop for calculate FDR, sorting P-value before this test

```python
estadistica2 = estadistica2.sort_values(by ='P',ascending=True).reset_index(drop=True)

estadistica2['Rank'] = estadistica2.index + 1
```

```python
estadistica2
```

# Ajustar el P value usando el método de Benjamini-Hochberg  (FDR)
## Definir el nivel de significancia o alfa

# Extraer los términos estadísticamente significativos

## <font color = purple>T = True if P <= FDR</font>
## <font color = purple>F = False if P > FDR</font>

<img src="https://raw.githubusercontent.com/Bioinformatica2020/Semana10/master/BENJAMINI_HOCHBERG.png" width = 80%>
    
https://github.com/Bioinformatica2020/Semana10/blob/master/Controlling_the_False_Discovery_Rate.pdf

## <font color = darkorange>Los métodos de corrección de pruebas múltiples reducen sistemáticamente las significancia de cada valo de P</font>
    
## <font color = darkorange>FDR es la probabilidad de que una hipótesis nula sea cierta habiendo sido rechazada por el test estadístico.</font>

```python
# 5% de falsos positivos
alfa = 0.05
```

```python
FDR_val=[]
for x in estadistica2.Rank:
    FDR_val.append((x/estadistica2.count()[0]) * alfa)
estadistica2['FDR'] = FDR_val
```

```python
estadistica2
```

# Loop para identificar GO terms significantes

```python
significant = []
for index, row in estadistica2.iterrows():
    if row.P <= row.FDR:
        significant.append('T')
    if row.P > row.FDR:
        significant.append('F')
estadistica2['Sig'] = significant
```
